# World Happiness Map — All Countries (2025)

Interactive choropleth for all 147 countries in the 2025 World Happiness Report.

| Tier | Score range | Countries |
|------|-------------|----------|
| Top | 6.69 - 7.76 | 30 |
| Upper-Mid | 6.01 - 6.64 | 40 |
| Lower-Mid | 4.67 - 6.01 | 40 |
| Bottom | 1.45 - 4.66 | 37 |

```python
try:
    import plotly
except ImportError:
    import subprocess, sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "plotly", "-q"])
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv("happiness_clean.csv")
df25 = df[df["year"] == 2025].copy()
df25["happiness_score"] = df25["happiness_score"].round(3)

print(f"Countries in 2025: {len(df25)}")
print(df25.groupby("tier")["happiness_score"].agg(["count", "min", "max"]))
```

```python
color_map = {
    "Top":        "#27ae60",
    "Upper-Mid":  "#a8d08d",
    "Lower-Mid":  "#e67e22",
    "Bottom":     "#c0392b",
}

fig = px.choropleth(
    df25,
    locations="country",
    locationmode="country names",
    color="tier",
    color_discrete_map=color_map,
    hover_name="country",
    hover_data={"happiness_score": True, "rank_in_year": True, "tier": False},
    category_orders={"tier": ["Top", "Upper-Mid", "Lower-Mid", "Bottom"]},
    title="<b>World Happiness Rating — All Countries (2025)</b>",
    labels={"rank_in_year": "Global rank", "happiness_score": "Score"},
)

fig.update_traces(marker_line_color="white", marker_line_width=0.5)

fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        coastlinecolor="#bdc3c7",
        showland=True,
        landcolor="#ecf0f1",
        showocean=True,
        oceancolor="#d6eaf8",
        showlakes=False,
        projection_type="natural earth",
    ),
    legend=dict(
        title="<b>Happiness Tier</b>",
        x=0.01, y=0.45,
        bgcolor="rgba(255,255,255,0.85)",
        bordercolor="#bdc3c7",
        borderwidth=1,
    ),
    margin=dict(l=0, r=0, t=60, b=0),
    width=1100, height=620,
    title_font_size=18,
    paper_bgcolor="white",
)

fig.show()
```

```python
# Horizontal bar chart — all 147 countries sorted by score
df_sorted = df25.sort_values("happiness_score")
bar_colors = [color_map[t] for t in df_sorted["tier"]]

fig2 = go.Figure(go.Bar(
    x=df_sorted["happiness_score"],
    y=df_sorted["country"],
    orientation="h",
    marker_color=bar_colors,
    hovertemplate="<b>%{y}</b><br>Score: %{x:.3f}<extra></extra>",
))

# Tier boundary lines
boundaries = [
    (df25[df25["tier"] == "Top"]["happiness_score"].min(), "Top"),
    (df25[df25["tier"] == "Upper-Mid"]["happiness_score"].min(), "Upper-Mid"),
    (df25[df25["tier"] == "Lower-Mid"]["happiness_score"].min(), "Lower-Mid"),
]
for x_val, label in boundaries:
    fig2.add_vline(
        x=x_val, line_dash="dash", line_color="#7f8c8d", line_width=1,
        annotation_text=label, annotation_position="top",
        annotation_font_size=9, annotation_font_color="#7f8c8d",
    )

fig2.update_layout(
    title="<b>Happiness Scores — All Countries (2025)</b>",
    xaxis_title="Happiness Score",
    xaxis_range=[0, 8.5],
    height=3200, width=860,
    plot_bgcolor="white", paper_bgcolor="white",
    xaxis=dict(showgrid=True, gridcolor="#ecf0f1"),
    yaxis=dict(tickfont=dict(size=10)),
    margin=dict(l=10, r=20, t=60, b=40),
    title_font_size=16,
)

fig2.show()
```
